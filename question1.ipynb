```python
import pandas as pd
import numpy as np
a = pd.read_csv('/content/Placement_BeginnerTask01.csv')
a = pd.get_dummies(a, drop_first=True).astype(float)
X = a.drop('PlacementStatus_Placed', axis = 1)
y = a['PlacementStatus_Placed']
```

```python
mean = X.mean(axis=0)
std = X.std(axis=0)
std[std==0] = 1.0
X = (X - mean)/std
train_size = 0.8 * len(X)
X_train = X[:int(train_size)]
y_train = y[:int(train_size)]
X_test = X[int(train_size):]
y_test = y[int(train_size):]
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def loss(y, y_pred):
  return -np.mean(y*np.log(y_pred + 1e-9)+(1-y)*np.log(1-y_pred + 1e-9))
```

```python
def train_logistic(X_train, y_train, alpha = 0.01, epochs=100):
  X_train = np.asarray(X_train)
  y_train = np.asarray(y_train)
  y_train = y_train.reshape(-1, 1)


  m, n = X.shape

  W = np.zeros((n, 1))
  b = 0

  for i in range(epochs):
    z = np.dot(X_train, W) + b
    y_pred = sigmoid(z)
    W = W - alpha*np.dot(X_train.T, (y_pred - y_train))/m
    b = b - alpha*np.mean(y_pred - y_train)
  return W, b
```

```python
W, b = train_logistic(X_train,y_train,epochs=10000)
```

```python
probs = sigmoid(np.dot(X_test,W)+b)
preds = (probs >= 0.5).astype(int)
y_test = np.asarray(y_test).reshape(-1, 1)
accuracy = np.mean(preds == y_test)
print(accuracy*100)
```
